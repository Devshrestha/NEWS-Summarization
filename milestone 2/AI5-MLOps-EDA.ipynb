```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme()
```

<h2>Data Cleaning Utils

```python
def read_csv_data(path):
    return pd.read_csv(path)

def clean_text(text):
    '''
    Just keeping alphabets and number
    and removing everything.
    '''
    newtext = re.sub(r"^[^\w]+|[^\w' ']", "", text)
    newtext = newtext.lower()
    return newtext

def data_overview(data):
    '''
    Giving a glance at the data
    ,Columns, columns type and Nans 
    information are displayed.
    '''
    display(data.head(2))
    print('+'*100)
    display(data.info())
    print('+'*100)
    print('TOTAL NaN')
    display(data.isna().sum())
    
    return None
    
def drop_nan(data):
    print('DROPPING NaNs')
    data = data.dropna()
    return data

def change_into_datetime(data):
    print('CHANGING INTO DATETIME TYPE FROM OBJECT')
    data['date'] = pd.to_datetime(data['date'])
    return data

def change_into_float(data):
    data['compression'] = data['compression'].astype('float')
    return data

def text_cleanig(data):
    data['title'] = data['title'].apply(lambda x : clean_text(x))
    data['text'] = data['text'].apply(lambda x : clean_text(x))
    return data

def data_cleaning(data_path):
    data = read_csv_data(data_path)
    data_overview(data)
    data = drop_nan(data)
    data = change_into_datetime(data)
    data = change_into_float(data)
    data = text_cleanig(data)
    
    print('DATA OVERVIEW AFTER CLEANING')
    data_overview(data)
    return data
```

<h2>Running the Pipeline.

```python
train1 = data_cleaning('train1.csv')
train2 = data_cleaning('train2.csv')
test = data_cleaning('test.csv')
```

<h2> Observing the Distributions

```python
train = pd.concat([train1,train2])
train.shape
```

```python
def Gaussian_Distribution(data,col):
    plt.figure(figsize=(12,8))
    plt.title(col,fontsize=20)
    sns.distplot(data[col])
    plt.show()
for col in ['compression','coverage','density'] :   
    Gaussian_Distribution(train,col)
```

<h3> We have skewed distribution for all the float type features.

<h2> Categorical features distribution

```python
def cat_dist(data,col):
    plt.figure(figsize=(24,8))
    data.groupby(col)[col].count().plot.bar()
    plt.show()

for col in ['compression_bin','coverage_bin','density_bin']:
    cat_dist(train,col)
```

<h3> As we can see categorical variables are uniformly distributed over the train dataset.

<h2>Time Series Analysis of float type features

```python
import numpy as np

def Time_Series_plot(data,col):
    plt.figure(figsize=(24,8))
    plt.title(col,fontsize=20)
    ind = np.argsort(data.date)
    x = data['date'].iloc[ind]
    y = data[col].iloc[ind]
    sns.scatterplot(data[col])
    plt.show()
    
for col in ['compression','coverage','density'] :   
    Time_Series_plot(train,col)
```
